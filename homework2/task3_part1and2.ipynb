```python
import pandas as pd
import openpyxl
import matplotlib.pyplot as plt
import seaborn as sn
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
import numpy as np
```

```python
df=pd.read_excel('Balance - Tablas (Parte 1 ).xlsx','pacientes',engine='openpyxl')
df.head()
```

```python
#extracting lst of all patients from balnce table part1 and storing it in patients dataframe
patients=df['IdPaciente']
patients
```

```python
# data=pd.read_excel('Balance - Tablas (Parte 1 ).xlsx','agenda',parse_dates=[['Fecha']],engine='openpyxl')
data=pd.read_excel('Balance - Tablas (Parte 1 ).xlsx','agenda',engine='openpyxl')
data.head()
```

```python
data.info()
data.columns
```

```python
data_1=data.iloc[:,1:4]
data_1.head()
```

```python
# data_1['Fecha'].to_datetime('2013-01-02 00:00:00', format='%Y%m%d', errors='ignore')
# print(data_1['Fecha'].dt.year)
```

```python
#replacing TODOS in dataframe with 1 and rest with zero
#TODOS means all service appointsments
arr=[]
for val in data_1['Area']:
    if val=='TODOS':
        arr.append(1)
    else:
        arr.append(0)
        
data_1['Area']=arr
data_1
```

```python
data_2=data_1[data_1['IdPaciente'].isnull()==False]
data_2['IdPaciente'].isnull().sum()
```

```python
# data_1['Fecha'] = pd.to_datetime(data_1['Fecha']).dt.date
# data_1
```

```python
data2 = pd.read_excel('IV_Datos_CMU.xlsx','cm_tickets',engine='openpyxl')
data2.head()
```

```python
data2['CompraTienda'].nunique()
```

```python
#patients who buy either medicines or products in dataframe df2
df2=data2.drop(columns='AbonoMembresiaInscripcion')
# arr=[]
# for i in range(len(df2)):
#     if ((df2[i]['CompraFarmacia']==0 )| (df2[i]['CompraFarmacia']==1)):
#         arr.append(1)
#     else:
#         arr.append(1)
# df2['buys']=arr
# df2=df2[df2.CompraFarmacia==1 | df2.CompraTienda==1]
df2['buys']=df2['CompraTienda']|df2['CompraFarmacia']
df2=df2[df2['buys']==1]
df2
```

```python
df2.info()
```

```python
#selecting patients who attended single service appointments: area=0
single_service=data_2[data_2["Area"]==0]
single_service
```

```python
fecha,hora=[],[]
absent_index=[]
for i in range(len(single_service)):
    try:
        f,h=(str(single_service.iloc[i,-1])).split(' ')
        fecha.append(f)
        hora.append(h)
    except:
        absent_index.append(i)
        continue
# single_service['Fecha']=pd.to_datetime(single_service['Fecha']).dt.date
# arr=[]
# single_service_1=single_service['Fecha'][:]
# for val in single_service['Fecha']:
#     arr.append(str(val)[:10])
        
# single_service['Fecha']=arr
single_service.info()
```

```python
print(len(fecha))
single_service=single_service.drop(columns=['Fecha'])
single_service=single_service.drop(absent_index)
single_service['Fecha']=fecha
single_service['hora']=hora
```

```python
single_service['Fecha'] = pd.to_datetime(single_service['Fecha'])
```

```python
single_service.info()
print(single_service.isnull().sum())
```

```python
# joining tables...who bought prodcuts in single service appointments
single_service_joined_data=pd.merge(single_service,df2,on=['IdPaciente','Fecha'])
single_service_joined_data.head()
```

```python
# joining this data with all patients
single_service_patients=pd.merge(patients,single_service_joined_data,on=['IdPaciente'],how='left')
single_service_patients.head()
```

```python
#replacing buys column null values with zer and converting it to value column
single_service_patients['buys']=single_service_patients['buys'].fillna(0)
single_service_patients=single_service_patients.rename(columns={'buys': 'value'})
single_service_patients
```

```python
single_service_patients=single_service_patients.drop(columns=['Area','hora','CompraFarmacia','CompraTienda'])
single_service_patients
```

```python
single_service_patients.to_csv('single_service_buys_something.csv',header=True)
```

```python
single_service_patients['value'].describe()
```
